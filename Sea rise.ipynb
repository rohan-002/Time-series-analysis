```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('sea rise.xlsx')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df.set_index('Date',inplace=True)
df.head()
```

```python
sns.set_theme(style = 'dark',font = 'monospace')
sns.lineplot(x = df.index,y=df['Value'],hue = df['Measure']).set(title='Annual rise in seal level')
plt.savefig('sealevel.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
timeseries = df['Value']
timeseries.rolling(100).mean().plot(label='100 year Rolling Mean')
#timeseries.rolling(100).std().plot(label='100 year Rolling Std')
timeseries.plot().set(title = 'Sea level Rise')
plt.legend
plt.ylabel('Sea level rise in mm')
plt.text("2016",-30,"Rohan Yadav")
plt.savefig('searise.png', dpi=300, bbox_inches='tight')
plt.show()
```
