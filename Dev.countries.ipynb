```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_excel('france.xlsx')
df1.head()
df2 = pd.read_excel('germany.xlsx')
df2.head()
df3 = pd.read_excel('switzerland.xlsx')
df3.head()
df4 = pd.read_excel('Uk.xlsx')
df4.head()
df5 = pd.read_excel('Us.xlsx')
df5.head()
```

```python
df1['year'] = pd.to_datetime(df1['year'],format = '%Y')
df1.head()
df2['year'] = pd.to_datetime(df2['year'],format = '%Y')
df2.head()
df3['year'] = pd.to_datetime(df3['year'],format = '%Y')
df4['year'] = pd.to_datetime(df4['year'],format = '%Y')
df5['year'] = pd.to_datetime(df5['year'],format = '%Y')
df5.head()
```

```python
df1.set_index('year',inplace=True)
df1.head()
df2.set_index('year',inplace=True)
df2.head()
df3.set_index('year',inplace=True)
df3.head()
df4.set_index('year',inplace=True)
df4.head()
df5.set_index('year',inplace=True)
df5.head()
```

```python
sns.set_theme(style = 'dark',font = 'monospace')
sns.lineplot(x = df1.index,y=df1['gdp'],color = 'red',label = 'France')
sns.lineplot(x = df2.index,y=df2['gdp'],color = 'green',label = 'Germany')
sns.lineplot(x = df4.index,y=df4['gdp'],color = 'black',label = 'United Kingdom')
sns.lineplot(x = df5.index,y=df5['gdp'],color = 'orange',label = 'United States').set(title= 'GDP of some Developed countries')
plt.xlabel("Year")
plt.ylabel("GDP")
plt.text(10000,0,"Rohan Yadav")
plt.savefig('GDPdev.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.lineplot(x = df1.index,y=df1['co2'],color = 'red',label = 'France' )
sns.lineplot(x = df2.index,y=df2['co2'],color = 'green',label = 'Germany')
sns.lineplot(x = df4.index,y=df4['co2'],color = 'black',label = 'United Kingdom')
sns.lineplot(x = df5.index,y=df5['co2'],color = 'orange',label = 'United States').set(title= 'Annual CO2 emissions')
plt.legend()
plt.xlabel("Years")
plt.ylabel("CO2 Emissions")
plt.text(10000,0,"Rohan Yadav")
plt.savefig('co2emissiondev.png', dpi=300, bbox_inches='tight')
plt.show()
```
